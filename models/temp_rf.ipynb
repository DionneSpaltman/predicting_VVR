```python
file_path_extracted_features = '/Users/dionnespaltman/Desktop/downloading_new/extracted_features_clean_incomplete.csv'

# Read the pickle file into a DataFrame
extracted_features_incomplete = pd.read_csv(file_path_extracted_features)

extracted_features_incomplete.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)

# Display the DataFrame
display(extracted_features_incomplete)

print(len(extracted_features_incomplete))
```

```python
# load the donor info file
donor_info_path = '/Users/dionnespaltman/Desktop/downloading_new/faint_info_final.csv'
donor_info = pd.read_csv(donor_info_path)
display(donor_info)

print(len(donor_info))
```

```python
# Filter IDs in donor_info based on extracted_features_incomplete
filtered_donor_info = donor_info[donor_info['ID'].isin(extracted_features_incomplete['ID'])]

print(len(filtered_donor_info))
```

```python
# Find IDs present in extracted_features_incomplete but missing in donor_info
missing_ids = extracted_features_incomplete[~extracted_features_incomplete['ID'].isin(donor_info['ID'])]['ID']

# Display the missing IDs
print("Missing IDs:", missing_ids)
```

```python
# Filter IDs in extracted_features_incomplete based on filtered_donor_info
filtered_extracted_features = extracted_features_incomplete[extracted_features_incomplete['ID'].isin(filtered_donor_info['ID'])]

# Now filtered_extracted_features contains only the rows with IDs present in filtered_donor_info
print(len(filtered_extracted_features))
```

## Merging files 

```python
# Merge filtered_donor_info and filtered_extracted_features on the 'ID' column
merged_df = pd.merge(filtered_donor_info, filtered_extracted_features, on='ID')

# Now merged_df contains both donor information and extracted features for the shared IDs

display(merged_df)
```

```python
print(merged_df.columns)
```

```python
# Create your X and y variables
# Assuming 'target_variable' is the column in merged_df indicating low or high VVR group
X = merged_df.drop(['ID', 'VVR_group', 'Unnamed: 0', 'Time_point', 'Gender', 'Age', 'Datum', 'Location',
       'Condition', 'Weight', 'Length', 'FEAR_sum', 'VVR_sum_tp', 'VVR_psych_tp', 'VVR_phys_tp', 'Faintness',
       'Dizziness', 'Weakness', 'Lightheadedness', 'Fear', 'Tension', 'Stress', 'Nervousness', 'Physical_sum', 'Psychological_sum',
       'Faintnessn', 'Dizzinessn', 'Weaknessn', 'Lightheadednessn', 'Fearn',
       'Tensionn', 'Stressn', 'Nervousnessn', 'na.rm', 'sumVVR_total', 'sumVVR_456'], axis=1)  # Features excluding 'ID' and target variable
y = merged_df['VVR_group']  # Target variable

# Now you can proceed with your random forest model using X and y
```

```python
print(X.columns)
print(len(X.columns))
```

## Running Random Forest 

```python
from sklearn.model_selection import train_test_split
# implementing train-test-split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=66)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=66)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report, confusion_matrix

# SMOTE toepassen om de dataset te balanceren
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Random Forest Classifier initialiseren en trainen met de gebalanceerde dataset
rf_smote = RandomForestClassifier(n_estimators=100, random_state=42)
rf_smote.fit(X_train_smote, y_train_smote)

# Voorspellingen doen op de testset
y_pred_smote = rf_smote.predict(X_test)
print(classification_report(y_test, y_pred_smote))
```

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier 

# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(X_train_resampled, y_train_resampled)

# predictions
rfc_predict = rfc.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rfc_cv_score = cross_val_score(rfc, X, y, cv=10, scoring='roc_auc')
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# number of features at every split
max_features = ['auto', 'sqrt']

# max depth
max_depth = [int(x) for x in np.linspace(100, 500, num = 11)]
max_depth.append(None)
# create random grid
random_grid = {
 'n_estimators': n_estimators,
 'max_features': max_features,
 'max_depth': max_depth
 }
# Random search of parameters
rfc_random = RandomizedSearchCV(estimator = rfc, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the model
rfc_random.fit(X_train, y_train)
# print results
print(rfc_random.best_params_)
```

```python
rfc = RandomForestClassifier(n_estimators=600, max_depth=300, max_features='sqrt')
rfc.fit(X_train,y_train)
rfc_predict = rfc.predict(X_test)
rfc_cv_score = cross_val_score(rfc, X, y, cv=10, scoring='roc_auc')
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```
