# 1. Data preparation

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.feature_selection import RFE
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
import matplotlib.pyplot as plt

# Load the data into pandas dataframes
X_train_res = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/X_train_res.csv')
y_train_res = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/y_train_res.csv')
X_test = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/X_test.csv')
y_test = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/y_test.csv')

# Drop the 'Unnamed: 0' column
X_train_res = X_train_res.drop(columns='Unnamed: 0', axis=1)
y_train_res = y_train_res.drop(columns='Unnamed: 0', axis=1)
X_test = X_test.drop(columns='Unnamed: 0', axis=1)
y_test = y_test.drop(columns='Unnamed: 0', axis=1)

# Convert the DataFrame to a 1-dimensional NumPy array
y_train_res = y_train_res.values.ravel()
y_test = y_test.values.ravel()

# Load gender-specific data
X_train_res_gender = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/X_train_res_gender.csv')
X_train_res_gender = X_train_res_gender.drop(columns='Unnamed: 0', axis=1)

X_test_gender = pd.read_csv('/Users/dionnespaltman/Desktop/V4/data/X_test_gender.csv')
X_test_gender = X_test_gender.drop(columns='Unnamed: 0', axis=1)

# Display the first few rows to verify loading
print("X_train_res:")
display(X_train_res.head())

print("\ny_train_res:")
display(y_train_res[:5])

print("\nX_test:")
display(X_test.head())

print("\ny_test:")
display(y_test[:5])

print("\nX_train_res_gender:")
display(X_train_res_gender.head())

print("\nX_test_gender:")
display(X_test_gender.head())
```

# 2. Define and Compare Multiple Base Models for RFE

```python
# Define the base models
models = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'SVM': SVC(probability=True, random_state=42),
    'NeuralNetwork': MLPClassifier(random_state=42, max_iter=300)
}

# Function to perform RFE with cross-validation and collect performance metrics
def evaluate_model_with_rfe(X, y, model_name, base_model, max_features, param_grid, cv_outer=5, cv_inner=5):
    scores = []
    feature_range = range(1, max_features + 1)
    
    for n_features in feature_range:
        # Update the number of features to select in RFE
        rfe = RFE(estimator=base_model, n_features_to_select=n_features, step=1)
        
        # Create a pipeline
        pipeline = Pipeline([
            ('rfe', rfe),
            ('classifier', base_model)
        ])
        
        # Inner loop: Grid search for hyperparameter tuning
        grid_search = GridSearchCV(pipeline, param_grid, cv=cv_inner, scoring='accuracy')
        
        # Outer loop: Cross-validation to estimate model performance
        nested_score = cross_val_score(grid_search, X, y, cv=cv_outer, scoring='accuracy')
        
        # Collect the mean score for this number of features
        scores.append(nested_score.mean())
    
    return feature_range, scores

# Define the parameter grids for hyperparameter tuning
param_grids = {
    'RandomForest': {'classifier__n_estimators': [50, 100, 200], 'classifier__max_depth': [None, 10, 20]},
    'DecisionTree': {'classifier__max_depth': [None, 10, 20]},
    'XGBoost': {'classifier__n_estimators': [50, 100, 200], 'classifier__max_depth': [3, 6, 10]},
    'SVM': {'classifier__C': [0.1, 1, 10], 'classifier__kernel': ['linear', 'rbf']},
    'NeuralNetwork': {'classifier__hidden_layer_sizes': [(50,), (100,), (50, 50)], 'classifier__alpha': [0.0001, 0.001]}
}

# Evaluate model performance with RFE for each base model
results = {}

for model_name, base_model in models.items():
    print(f"Evaluating {model_name}...")
    param_grid = param_grids[model_name]
    max_features = X_train_res.shape[1]
    feature_range, scores = evaluate_model_with_rfe(X_train_res, y_train_res, model_name, base_model, max_features, param_grid)
    results[model_name] = (feature_range, scores)

# Print scores for each model and each number of features
for model_name, (feature_range, scores) in results.items():
    print(f"\nModel: {model_name}")
    for n_features, score in zip(feature_range, scores):
        print(f'Number of features: {n_features}, Score: {score}')
```

# 3. Plot Performance vs. Number of Features for Each Model

```python
plt.figure(figsize=(14, 8))

for model_name, (feature_range, scores) in results.items():
    plt.plot(feature_range, scores, marker='o', label=model_name)

plt.xlabel('Number of Features')
plt.ylabel('Accuracy')
plt.title('Model Performance vs. Number of Features')
plt.legend()
plt.grid(True)
plt.show()
```
