```python
# Step 1: Group the DataFrame by 'ID'
grouped = toy_data.groupby('ID')

# Step 2: Calculate the sum of the values of the 8 new variables for each group
sum_8_vars = grouped[['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum()

# Step 3: Create a new DataFrame with the sum of the 8 variables and other values copied from the row with timepoint 1.0
extra_rows = pd.DataFrame(columns=toy_data.columns)
for name, group in grouped:
    extra_row = group[group['Time_point'] == 1.0].copy()
    extra_row['Time_point'] = -1.0
    extra_row['Fear'] = sum_8_vars.loc[name]['Fear']
    extra_row['Stress'] = sum_8_vars.loc[name]['Stress']
    extra_row['Tension'] = sum_8_vars.loc[name]['Tension']
    extra_row['Nervousness'] = sum_8_vars.loc[name]['Nervousness']
    extra_row['Faintness'] = sum_8_vars.loc[name]['Faintness']
    extra_row['Dizziness'] = sum_8_vars.loc[name]['Dizziness']
    extra_row['Weakness'] = sum_8_vars.loc[name]['Weakness']
    extra_row['Lightheadedness'] = sum_8_vars.loc[name]['Lightheadedness']
    extra_rows = extra_rows.append(extra_row.iloc[0], ignore_index=True)  # Append only the first row for each ID

# Step 4: Print the new DataFrame
display(extra_rows)
```

```python
toy_and_extra = pd.concat([toy_data, extra_rows])

display(toy_and_extra)
```

What I want, is working. So now it's time to add two extra columns sumVVR_456 and sumVVR_total (last one is to double check). Only if the Time_point is -1.0, I want the sum of the 8 variables. 

```python
import numpy as np

# Create the new columns sumVVR_456 and sumVVR_total
toy_and_extra['sumVVR_456_temp'] = np.zeros(len(toy_and_extra))
toy_and_extra['sumVVR_total'] = toy_and_extra[['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

# Calculate the sum of the 8 variables if Time_point is -1.0
toy_and_extra.loc[toy_and_extra['Time_point'] == -1.0, 'sumVVR_456'] = \
    toy_and_extra.loc[toy_and_extra['Time_point'] == -1.0, ['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

# Calculate sumVVR_456 for timepoints 4, 5, and 6
toy_and_extra.loc[toy_and_extra['Time_point'].isin([4.0, 5.0, 6.0]), 'sumVVR_456_temp'] = \
    toy_and_extra.loc[toy_and_extra['Time_point'].isin([4.0, 5.0, 6.0]), ['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

sum_per_id = toy_and_extra[toy_and_extra['Time_point'] != -1.0].groupby('ID')['sumVVR_456_temp'].sum()
toy_and_extra['sumVVR_456'] = toy_and_extra.apply(lambda row: sum_per_id[row['ID']] if row['Time_point'] == -1.0 else row['sumVVR_456_temp'], axis=1)

toy_and_extra.drop(columns=['sumVVR_456_temp'], inplace=True)

# Display the updated DataFrame
display(toy_and_extra)
```

### Adding the sum (to faint_info)

```python
# Step 1: Group the DataFrame by 'ID'
grouped_faint_info = faint_info.groupby('ID')

# Step 2: Calculate the sum of the values of the 8 new variables for each group
sum_8_vars = grouped_faint_info[['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum()

# Step 3: Create a new DataFrame with the sum of the 8 variables and other values copied from the row with timepoint 1.0
extra_rows_faint_info = pd.DataFrame(columns=faint_info.columns)
for name, group in grouped_faint_info:
    extra_row = group[group['Time_point'] == 1.0].copy()
    extra_row['Time_point'] = -1.0
    extra_row['Fear'] = sum_8_vars.loc[name]['Fear']
    extra_row['Stress'] = sum_8_vars.loc[name]['Stress']
    extra_row['Tension'] = sum_8_vars.loc[name]['Tension']
    extra_row['Nervousness'] = sum_8_vars.loc[name]['Nervousness']
    extra_row['Faintness'] = sum_8_vars.loc[name]['Faintness']
    extra_row['Dizziness'] = sum_8_vars.loc[name]['Dizziness']
    extra_row['Weakness'] = sum_8_vars.loc[name]['Weakness']
    extra_row['Lightheadedness'] = sum_8_vars.loc[name]['Lightheadedness']
    extra_rows_faint_info = extra_rows_faint_info.append(extra_row.iloc[0], ignore_index=True)  # Append only the first row for each ID

# # Step 4: Print the new DataFrame
# display(extra_rows_faint_info)

faint_info_and_extra = pd.concat([faint_info, extra_rows_faint_info])

display(extra_rows_faint_info)
```

```python
import numpy as np

# Create the new columns sumVVR_456 and sumVVR_total
faint_info_and_extra['sumVVR_456_temp'] = np.zeros(len(faint_info_and_extra))
faint_info_and_extra['sumVVR_total'] = faint_info_and_extra[['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

# Calculate the sum of the 8 variables if Time_point is -1.0
faint_info_and_extra.loc[faint_info_and_extra['Time_point'] == -1.0, 'sumVVR_456'] = \
    faint_info_and_extra.loc[faint_info_and_extra['Time_point'] == -1.0, ['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

# Calculate sumVVR_456 for timepoints 4, 5, and 6
faint_info_and_extra.loc[faint_info_and_extra['Time_point'].isin([4.0, 5.0, 6.0]), 'sumVVR_456_temp'] = \
    faint_info_and_extra.loc[faint_info_and_extra['Time_point'].isin([4.0, 5.0, 6.0]), ['Fear', 'Stress', 'Tension', 'Nervousness', 'Faintness', 'Dizziness', 'Weakness', 'Lightheadedness']].sum(axis=1)

sum_per_id = faint_info_and_extra[faint_info_and_extra['Time_point'] != -1.0].groupby('ID')['sumVVR_456_temp'].sum()
faint_info_and_extra['sumVVR_456'] = faint_info_and_extra.apply(lambda row: sum_per_id[row['ID']] if row['Time_point'] == -1.0 else row['sumVVR_456_temp'], axis=1)

faint_info_and_extra.drop(columns=['sumVVR_456_temp'], inplace=True)

# Display the updated DataFrame
display(faint_info_and_extra)
```
