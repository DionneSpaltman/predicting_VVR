Import statements and data 

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
filtered_donor_info = pd.read_csv('/Users/dionnespaltman/Desktop/V6/filtered_donor_info_03-06-2024.csv', sep=',')

if 'Unnamed: 0' in filtered_donor_info.columns:
    filtered_donor_info.drop(columns=['Unnamed: 0'], inplace=True)

display(filtered_donor_info)
```

# Get the range of dates 

```python
# Find the first and last dates
first_date = filtered_donor_info['Date'].min()
last_date = filtered_donor_info['Date'].max()

print("First date:", first_date)
print("Last date:", last_date)
```

# Counts 

```python
# Ensure each ID is unique by dropping duplicates based on 'ID'
unique_donor_info = filtered_donor_info.drop_duplicates(subset='ID')

# Count occurrences of VVR_group
vvr_group_counts = unique_donor_info['VVR_group'].value_counts()
print("\nVVR_group counts (per unique ID):\n", vvr_group_counts)

# Count occurrences of Condition
condition_counts = unique_donor_info['Condition'].value_counts()
print("\nCondition counts (per unique ID):\n", condition_counts)

# Count occurrences of Gender
gender_counts = unique_donor_info['Gender'].value_counts()
print("\nGender counts (per unique ID):\n", gender_counts)
```
