# Descriptives of the 104 participants 

# Import statements and data 

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
complete_data = pd.read_csv('/Users/dionnespaltman/Desktop/V4/filtered_donor_info_104.csv', sep=',')

if 'Unnamed: 0' in complete_data.columns:
    complete_data.drop(columns=['Unnamed: 0'], inplace=True)


complete_data.rename(columns={'sum_VVR': 'Sum_VVR'}, inplace=True)

display(complete_data)
```

# Get the range of dates 

```python
# Find the first and last dates
first_date = complete_data['Date'].min()
last_date = complete_data['Date'].max()

print("First date:", first_date)
print("Last date:", last_date)
```

# VVR group needs to be created 

First we create the sum of stages 4, 5 and 6. 

```python
import pandas as pd

# Filter rows where Stage is 4, 5 or 6
VVR_measurements_filtered = complete_data[complete_data['Stage'].isin([4, 5, 6])]

# Group by ID and calculate the sum of Sum_VVR
sum_456 = VVR_measurements_filtered.groupby('ID')['Sum_VVR'].sum().reset_index()

# Merge sum_456 back into complete_data on 'ID'
complete_data = complete_data.merge(sum_456, on='ID', how='left', suffixes=('', '_456'))

# Rename the merged column for clarity
complete_data.rename(columns={'Sum_VVR_456': 'Sum_456'}, inplace=True)

# Get the maximum and minimum values of the sum_VVR column
max_sum_VVR = complete_data['Sum_456'].max()
min_sum_VVR = complete_data['Sum_456'].min()

print("Maximum value of Sum_456:", max_sum_VVR)
print("Minimum value of Sum_456:", min_sum_VVR)

# Calculate the mean of the 'Sum_456' column, excluding NaN values
mean_sum_456 = complete_data['Sum_456'].mean(skipna=True)

# Print the mean value
print("Mean of 'Sum_456' column:", mean_sum_456)

# Display the updated DataFrame
display(complete_data.head(20))
```

Now do the same but for stages 4, 5, 6 and 7. 

```python
# Filter rows where Stage is 4, 5 6 or 7 
VVR_measurements_filtered = complete_data[complete_data['Stage'].isin([4, 5, 6, 7])]

# Group by ID and calculate the sum of Sum_VVR
sum_4567 = VVR_measurements_filtered.groupby('ID')['Sum_VVR'].sum().reset_index()

# Merge sum_456 back into complete_data on 'ID'
complete_data = complete_data.merge(sum_4567, on='ID', how='left', suffixes=('', '_4567'))

# Rename the merged column for clarity
complete_data.rename(columns={'Sum_VVR_4567': 'Sum_4567'}, inplace=True)

# Get the maximum and minimum values of the sum_VVR column
max_sum_VVR = complete_data['Sum_4567'].max()
min_sum_VVR = complete_data['Sum_4567'].min()

print("Maximum value of Sum_4567:", max_sum_VVR)
print("Minimum value of Sum_4567:", min_sum_VVR)

# Display the updated DataFrame
display(complete_data)
```

```python
# Calculate the mean of the 'Sum_456' column, excluding NaN values
mean_sum_4567 = complete_data['Sum_4567'].mean(skipna=True)

# Print the mean value
print("Mean of 'Sum_4567' column:", mean_sum_4567)

# Create a new column 'VVR_group' based on the condition
complete_data['VVR_group'] = complete_data['Sum_4567'].apply(lambda x: 0 if pd.isna(x) or x < mean_sum_4567 else 1)

# Display the updated DataFrame
display(complete_data)
```

In addition, I want to get the sum VVR of stages 1 and 2. 

```python
# Filter rows where Stage is 1 or 2 
VVR_measurements_filtered = complete_data[complete_data['Stage'].isin([1, 2])]

# Group by ID and calculate the sum of Sum_VVR
sum_12 = VVR_measurements_filtered.groupby('ID')['Sum_VVR'].sum().reset_index()

# Merge sum_456 back into complete_data on 'ID'
complete_data = complete_data.merge(sum_12, on='ID', how='left', suffixes=('', '_12'))

# Rename the merged column for clarity
complete_data.rename(columns={'Sum_VVR_12': 'Sum_12'}, inplace=True)

# Get the maximum and minimum values of the sum_VVR column
max_sum_VVR = complete_data['Sum_12'].max()
min_sum_VVR = complete_data['Sum_12'].min()

print("Maximum value of Sum_12:", max_sum_VVR)
print("Minimum value of Sum_12:", min_sum_VVR)

# Display the updated DataFrame
display(complete_data)
```

Also do it seperately for stage 1 and then for stage 2. 

```python
# Filter rows where Stage is 1 
VVR_measurements_filtered = complete_data[complete_data['Stage'].isin([1])]

# Group by ID and calculate the sum of Sum_VVR
sum_1 = VVR_measurements_filtered.groupby('ID')['Sum_VVR'].sum().reset_index()

# Merge sum_456 back into complete_data on 'ID'
complete_data = complete_data.merge(sum_1, on='ID', how='left', suffixes=('', '_1'))

# Rename the merged column for clarity
complete_data.rename(columns={'Sum_VVR_1': 'Sum_1'}, inplace=True)

# Get the maximum and minimum values of the sum_VVR column
max_sum_VVR = complete_data['Sum_1'].max()
min_sum_VVR = complete_data['Sum_1'].min()

print("Maximum value of Sum_1:", max_sum_VVR)
print("Minimum value of Sum_1:", min_sum_VVR)

# Display the updated DataFrame
display(complete_data)
```

```python
# Filter rows where Stage is 2
VVR_measurements_filtered = complete_data[complete_data['Stage'].isin([2])]

# Group by ID and calculate the sum of Sum_VVR
sum_2 = VVR_measurements_filtered.groupby('ID')['Sum_VVR'].sum().reset_index()

# Merge sum_456 back into complete_data on 'ID'
complete_data = complete_data.merge(sum_2, on='ID', how='left', suffixes=('', '_2'))

# Rename the merged column for clarity
complete_data.rename(columns={'Sum_VVR_2': 'Sum_2'}, inplace=True)

# Get the maximum and minimum values of the sum_VVR column
max_sum_VVR = complete_data['Sum_2'].max()
min_sum_VVR = complete_data['Sum_2'].min()

print("Maximum value of Sum_2:", max_sum_VVR)
print("Minimum value of Sum_2:", min_sum_VVR)

# Display the updated DataFrame
display(complete_data)
```

```python
# Drop duplicate IDs, keeping the first occurrence of each ID
unique_ids_data = complete_data.drop_duplicates(subset='ID')

# Count the occurrences of each unique value in the 'VVR_group' column
vvr_group_counts = unique_ids_data['VVR_group'].value_counts()

# Print the counts
print("Count of unique IDs in each 'VVR_group':")
print(vvr_group_counts)
```
